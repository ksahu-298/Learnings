## To create a line graph b/w X and Y axis ##


```python
import matplotlib.pyplot as plt

x= [1,2,3,4]
y=[10,20,15,25]

plt.plot(x,y)
plt.show()
```

in the above graph
    1 = 10
    2 = 20
    3 = 15
    4 = 25

## creating a simple line graph which uses all the basic functions of pyplot ##

```python
import matplotlib.pyplot as plt
x= ['Mon','Tue', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']  # x-axis
y = [10, 15, 5, 20 ,12, 0, 88] # y- axis

plt.plot(x,y,color = 'Red', linestyle='--', linewidth = 1, marker = '+', label = "Bakery Sales")
plt.title ("Weekly Bakery Cake Sale")
plt.xlabel("Weekdays")
plt.ylabel("Sales of Cakes")
plt.legend()
plt.show()
```
